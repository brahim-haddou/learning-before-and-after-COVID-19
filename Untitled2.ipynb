```python
df_model = df.drop([ 'Grade_level','LH_w_Instruction','onl_instr','onl_noninstr','Total_offline', 'off_instr', 'off_noninstr','nec_habit', 'nec_parent','kno_covid',  'kno_soci'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
X = df_model.values
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
from sklearn.cluster import KMeans
cs=[]
for i in range(1,10):
    k_means = KMeans(init = "k-means++", n_clusters = i)
    k_means.fit(Clus_dataSet)
    cs.append(k_means.inertia_)
    
plt.plot(range(1,10),cs)
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("CS")
plt.show()
```

```python
k_means = KMeans(init = "k-means++", n_clusters = 3)
k_means.fit(Clus_dataSet)
labels = k_means.labels_
df["Clus_km"] = labels
df['Clus_km'].value_counts().to_frame()
```

```python
df['Clus_km'] = df['Clus_km'].astype('category') 
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Clus_km', axis = 1)
y = df['Clus_km']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
Tree = DecisionTreeClassifier(max_depth = 3)
Tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred = Tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
