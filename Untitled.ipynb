# Student Learning Habit Before And During COVID-19
analyse, visualisation et regroupement les élèves en fonction des habitudes d'apprentissage avant et pendant le COVID-19

## About the dataset

The dataset was constructed to examine Vietnamese student’s learning habit during the school suspension time due to the novel coronavirus (Covid-19), responds to the call of open research to prevent potential effects of coronavirus . The questionnaires were spread over a network of educational communities on Facebook from 7th to 28th of February 2020. Using the snowball sampling method, researchers delivered the survey to teachers and parents to confirm the consent form before they forward it to their students and children. In order to measure the influence of student’s socioeconomic status and occupational aspirations over their learning habit during school closure, the survey includes three major groups of questions: (1) Individual demographic, includes family socioeconomic status, school types, and occupational aspirations; (2) Student’s learning habits, include hours of learning before and during school suspension time, with and without other people’s support; and (3) Student’s perception on self-learning during the disease. The dataset includes 420 valid observations.

### variables :

- **Gender** : (1) Male (2) Female (3) Not public
- **Class**  : grade of the student

- **Sib** : the number of siblings  (1) One (2) Two (3) Three (4) Over four
- **fa_job** : Father's job (1) Stem (2) Social siences (3) Free (4) Others
- **mo_job** : Mother's job (1) Stem (2) Social siences (3) Free (4) Others
- **income** : the monthly income of the family (1) Less than 430 USD (2) From 430 to under 860 USD (3) From 860 to under 1,290 USD (4)From 1,290 to under 1,720 USD (5) From 1,720 to under 2,150 USD (6) More than 2,150 USD

- **English** : student evaluate his English capability
- **Lh_before_Cov** : self-learning hours before COVID-19 (1) under 4h (2) from 4 to 7h (3) over 7h 
- **Lh_in_Cov** : self-learning hours during COVID-19     (1) under 4h (2) from 4 to 7h (3) over 7h

- **LH_w_Instruction** : Total hours of learning hours without instruction
- **onl_instr** : Total hours of online learning hours with instruction
- **onl_noninstr** : Total hours of online learning hours without instruction
- **Total_offline** : Regarding hours of learning during COVID-19, number of hours are offline learning
- **off_instr** : Total hours of offline learning hours with instruction
- **off_noninstr** : Total hours of offline learning hours without instruction

```python
gender ={
    1 : "Male",
    2 : "Female",
    3 : "Not public"
}
df["Gender"].replace(gender, inplace=True)

Class = {
    1 : 6,
    2 : 7,
    3 : 8,
    4 : 9,
    5 : 10,
    6 : 11,
    7 : 12
}
df["Class"].replace(Class, inplace=True)

Grade_level = {
    1 : "Secondary school",
    2 : "High school"
}

df["Grade_level"].replace(Grade_level, inplace=True)

school_type = {
    1 : "Public school (normal)",
    2 : "Public school (Gifted)",
    3 : "Private school (normal)",
    4 : "International school"
}
df["school_type"].replace(school_type, inplace=True)

sib = {
    1 : "One",
    2 : "Two",
    3 : "Three",
    4 : "Over four"
}
df["Sib"].replace(sib, inplace=True)

job = {
    1 : "Stem",
    2 : "Social siences",
    3 : "Free",
    4 : "Others"
}
df["fa_job"].replace(job, inplace=True)
df["mo_job"].replace(job, inplace=True)

income = {
    1 : "Less than 430 USD",
    2 : "From 430 to under 860 USD",
    3 : "From 860 to under 1,290 USD",
    4 : "Form 1,290 to under 1,720 USD",
    5 : "Form 1,720 to under 2,150 USD",
    6 : "More than 2,150 USD"
}
df["income"].replace(income , inplace=True)

exam = {
    1 : "A (Mathematics, Physics, Chemistry)",
    2 : "A1 (Mathematics, Physics, English)",
    3 : "B (Mathematics, Biology, Chemistry)",
    4 : "C (Literature, History, Geography)",
    5 : "D (Literature, Foreign Language, Mathematics)",
    6 : "Others"
}

df["exam"].replace(exam , inplace=True)

evaluation = {
    1 : "Below Average",
    2 : "Average",
    3 : "Good",
    4 : "Excellence"
}
df["Self_evaluation"].replace(evaluation , inplace=True)
df["English"].replace(evaluation , inplace=True)

learning_hours={
    1 : "under 4h",
    2 : "from 4 to 7h",
    3 : "over 7h"
}

df["Lh_before_Cov"].replace(learning_hours , inplace=True)
df["Lh_in_Cov"].replace(learning_hours , inplace=True)

opinion = {
    1 : "Strongly disagree",
    2 : "Disagree",
    3 : "Neither agree nor disagree",
    4 : "Agree",
    5 : "Strongly agree"
}

df['nec_prog'].replace(opinion , inplace=True) 
df['nec_habit'].replace(opinion , inplace=True) 
df['nec_eacher'].replace(opinion , inplace=True)
df['nec_parent'].replace(opinion , inplace=True)
df['nec_sib'].replace(opinion , inplace=True) 
df['nec_friend'].replace(opinion , inplace=True) 
df['eff_moti'].replace(opinion , inplace=True) 
df['eff_con'].replace(opinion , inplace=True) 
df['eff_supp'].replace(opinion , inplace=True) 
df['eff_env'].replace(opinion , inplace=True) 
df['eff_obj'].replace(opinion , inplace=True) 
df['eff_resource'].replace(opinion , inplace=True) 
df['eff_friend'].replace(opinion , inplace=True)
df['kno_med'].replace(opinion , inplace=True) 
df['kno_covid'].replace(opinion , inplace=True) 
df['kno_env'].replace(opinion , inplace=True) 
df['kno_soci'].replace(opinion , inplace=True) 
df['kno_elearn'].replace(opinion , inplace=True)
```
